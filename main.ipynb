# Training Auto Encoder

```python
from Autoencoder import layers, autoencoder, callbacks, dataset, trainloop

import matplotlib.pyplot as plt 
import tensorflow as tf 
```

```python
def main(model, train_ds, validation_ds, loss, optimizer, model_ckpt, history_log):
    train = trainloop.Train(model, loss, optimizer, model_ckpt, history_log)
    trained_model, history = train.trainLoop(train_ds, validation_ds, debug = True)
    
    return trained_model, history 
```

```python
import tensorflow as tf

if __name__ == "__main__":
    
    (X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
    
        
    ds = dataset.Dataset(X_train)
    train_ds, validation_ds = ds.dataset(True)
    plt.show() 

    model = autoencoder.AutoEncoder()
    loss = layers.Losses()
    optimizer = tf.keras.optimizers.Adam()
    model_ckpt = "ckpt"
    history_log = "history"
    
    trained_model, history = main(model, train_ds, validation_ds, loss, optimizer, model_ckpt, history_log)
    
#     trained_model.save_weights("weights/final.tf", save_format = "tf")
```

```python
import random 
import numpy as np 
import cv2 

def sample(x):
    sampled = np.random.randint(0, x.shape[0], size = 30)
    images = x[sampled]
        
    plt.figure(figsize=(20,20))
    columns = 10
    for i, image in enumerate(images):
        blur = cv2.GaussianBlur(image, (3, 3), 0)
        image = model(blur.reshape(-1, 28, 28, 1).astype("float32") / 255.0, pipeline = "autoencoder", training = False)
        ax = plt.subplot(len(images) / columns + 1, columns, i + 1)
        plt.axis('off')
        plt.subplots_adjust(bottom=0.1)
        plt.imshow(image.numpy().reshape(28, 28), cmap='gray')
```

# Training Classifcation Model

```python
import pathlib 
import numpy as np 
import cv2 

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
class FeatureExtraction(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.model = autoencoder.AutoEncoder()
        root = tf.train.Checkpoint(net = self.model)
        status = root.restore(tf.train.latest_checkpoint(pathlib.Path("ckpt"))).expect_partial()
    
    def fit(self, X, y = None):
        return self 
    
    def transform(self, X):
        X = X.copy()
        for i in range(len(X)):
            X[i] = cv2.GaussianBlur(X[i], (3, 3), 1)
    
        X = X.reshape(-1, 28, 28, 1).astype("float32") / 255.0    
        extracted_feature = self.model(X, pipeline = "classification", training = False)
        
        return extracted_feature.numpy()
```

```python
clf = Pipeline([
    ("features_extraction", FeatureExtraction()),
    ("classifier", RandomForestClassifier())
])
```

```python
for params in clf.get_params().keys():
    if "classifier" in params:
        print(params)
```

```python
params = {"classifier__max_depth": np.arange(10, 300, 10),
          "classifier__max_features": np.arange(10, 150, 5),
          "classifier__n_estimators": np.arange(20, 500, 5),
          "classifier__min_samples_split": np.linspace(0.01, 0.2, 10, endpoint = True),
         "classifier__max_features": ["auto", "sqrt", "log2"]
         }

randomized_search = RandomizedSearchCV(clf, params, n_iter = 10, cv = 3, random_state = 42,
                                      verbose = 2)
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

randomized_search.fit(X_train, y_train)
```
